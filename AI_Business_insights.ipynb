```python
# src/data_processor.py
import pandas as pd
import numpy as np

def load_and_unify() -> pd.DataFrame:
    # Load your exact files
    google = pd.read_csv("data/raw/google_ads-sheet1-sourcetable.csv")
    linkedin = pd.read_csv("data/raw/linked_in_campaigns-sheet1-sourcetable.csv")
    meta = pd.read_csv("data/raw/meta_ads_campigns-sheet1-sourcetable.csv")

    # CLEAN CURRENCY (you already did this  keeping it)
    google['Cost'] = google['Cost'].replace(r'[$,]', '', regex=True).astype(float)
    linkedin['Total Spent (USD)'] = linkedin['Total Spent (USD)'].replace(r'[$,]', '', regex=True).astype(float)
    if 'Cost per Lead (USD)' in linkedin.columns:
        linkedin['Cost per Lead (USD)'] = linkedin['Cost per Lead (USD)'].replace(r'[$,]', '', regex=True).astype(float)
    meta['Amount spent'] = meta['Amount spent'].replace(r'[$,]', '', regex=True).astype(float)
    if 'Purchase value' in meta.columns:
        meta['Purchase value'] = meta['Purchase value'].replace(r'[$,]', '', regex=True).astype(float)

    # STANDARDIZE (you already started  we finish it perfectly)
    google = google.rename(columns={
        'Day': 'Date', 'Campaign': 'Campaign Name', 'Country': 'Region', 'Cost': 'Spend'
    })
    google['Platform'] = 'Google Ads'
    google['Revenue'] = 0
    google['Conversions'] = google.get('Conversions', 0)

    linkedin = linkedin.rename(columns={
        'Day': 'Date', 'Total Spent (USD)': 'Spend'
    })
    if 'Conversions' in linkedin.columns:
        linkedin = linkedin.drop(columns=['Conversions'])
    linkedin = linkedin.rename(columns={'Leads': 'Conversions'})
    linkedin['Platform'] = 'LinkedIn'
    linkedin['Revenue'] = 0
    linkedin['Region'] = 'Global'
    linkedin['Device'] = 'Unknown'

    meta = meta.rename(columns={
        'Campaign name': 'Campaign Name', 'Amount spent': 'Spend',
        'Purchases': 'Conversions', 'Purchase value': 'Revenue'
    })
    meta['Platform'] = 'Meta'
    meta['Region'] = meta.get('Country', 'Unknown')
    meta['Device'] = 'Mixed'
    meta['Revenue'] = meta['Revenue'].fillna(0)

    # FINAL COLUMNS
    final_cols = ['Date', 'Platform', 'Campaign Name', 'Region', 'Device',
                  'Impressions', 'Clicks', 'Spend', 'Conversions', 'Revenue']

    google = google.reindex(columns=final_cols, fill_value=0)
    linkedin = linkedin.reindex(columns=final_cols, fill_value=0)
    meta = meta.reindex(columns=final_cols, fill_value=0)

    # COMBINE (you already did this!)
    df = pd.concat([google, linkedin, meta], ignore_index=True)
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # CALCULATE METRICS
    df['CTR'] = df['Clicks'] / df['Impressions'].replace(0, np.nan)
    df['CPC'] = df['Spend'] / df['Clicks'].replace(0, np.nan)
    df['ROAS'] = df['Revenue'] / df['Spend'].replace(0, np.nan)
    df['ROAS'] = df['ROAS'].fillna(0).replace([np.inf, -np.inf], 0)

    return df
```
